```python
table=pd.read_csv('Table.csv')
table.head()
```

#### Most matches winnig teams

```python
table[['TEAM','W']][table['W']==table['W'].max()]
```

```python
sns.barplot(x='TEAM', y='W', data=table, hue='TEAM', dodge=False,palette='Set2')
plt.ylabel('Total Wins')
for index, row in table.iterrows():
    plt.text(x=index, y=row['W'] , s=str(row['W']), ha='center', va='bottom')
plt.title("Team vs Total wins")
plt.show()
```

#### Most matches lossing team 

```python
table[['TEAM','L']][table['L']==table['L'].max()]
```

```python
sns.barplot(x='TEAM',y='L',data=table,hue='TEAM',palette='Set2')
for index,row in table.iterrows():
  plt.text(x=index,y=row['L'],s=str(row['L']),ha='center',va='bottom')
plt.ylabel('Total loses')
plt.title('Team Vs Total loss Matches')
plt.show()
```

#### Highest and lowest aggregate runs scored by team in season 25

```python
print('Most Scored Runs')
table[['TEAM','RUNS_FOR','OVERS_FOR']][table['RUNS_FOR']==table['RUNS_FOR'].max()]
```

```python
print('Least Scored Runs')
table[['TEAM','RUNS_FOR','OVERS_FOR']][table['RUNS_FOR']==table['RUNS_FOR'].min()]
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1,2,1)
sns.barplot(x='TEAM',y='RUNS_FOR',data=table,hue='TEAM',palette='Set2')
for index,row in table.iterrows():
  plt.text(x=index,y=row['RUNS_FOR'],s=str(row['RUNS_FOR']),ha='center',va='bottom')
plt.ylabel('Total Scored runs in season')
plt.title('Team Vs Total Aggregate Scored runs in season')

plt.subplot(1,2,2)
sns.barplot(x='TEAM',y='OVERS_FOR',data=table,hue='TEAM',palette='Set2')
for index,row in table.iterrows():
  plt.text(x=index,y=row['OVERS_FOR'],s=str(row['OVERS_FOR']),ha='center',va='bottom')
plt.ylabel('Total Overs Played in season')
plt.title('Team Vs Total Overs Played in season')

plt.tight_layout()
plt.show()
```

#### Highest and lower aggregate scored by opponent teams in season 25 

```python
print('Most Scored Runs')
table[['TEAM','RUNS_AGAINST','OVERS_AGAINST']][table['RUNS_AGAINST']==table['RUNS_AGAINST'].max()]
```

```python
print('least Scored Runs')
table[['TEAM','RUNS_AGAINST','OVERS_AGAINST']][table['RUNS_AGAINST']==table['RUNS_AGAINST'].min()]
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1,2,1)
sns.barplot(x='TEAM',y='RUNS_AGAINST',data=table,hue='TEAM',palette='Set2')
for index,row in table.iterrows():
  plt.text(x=index,y=row['RUNS_AGAINST'],s=str(row['RUNS_AGAINST']),ha='center',va='bottom')
plt.ylabel('Total Conceded runs in season')
plt.title('Team Vs Total Conceded runs in season')

plt.subplot(1,2,2)
sns.barplot(x='TEAM',y='OVERS_AGAINST',data=table,hue='TEAM',palette='Set2')
for index,row in table.iterrows():
  plt.text(x=index,y=row['OVERS_AGAINST'],s=str(row['OVERS_AGAINST']),ha='center',va='bottom')
plt.ylabel('Total Overs Bowled in season')
plt.title('Team Vs Total Overs Bowled in season')

plt.tight_layout()
plt.show()
```

```python
df = table.melt(id_vars='TEAM', 
                value_vars=['RUNS_FOR', 'RUNS_AGAINST'],
                var_name='Type', 
                value_name='Runs')
df['Type'] = df['Type'].map({
    'RUNS_FOR': 'Scored by Team',
    'RUNS_AGAINST': 'Scored by Opponent'
})
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='TEAM', y='Runs', hue='Type')
plt.title('Runs Scored vs Conceded by Each Team (Season 25)')
plt.tight_layout()
plt.show()
```

```python
sorted_df = table.sort_values(by='NRR')
plt.figure(figsize=(12, 6))
plt.plot(sorted_df['TEAM'], sorted_df['NRR'], marker='o', linestyle='-', color='green')
plt.title('Net Run Rate (NRR) of Teams - Season 25')
plt.xlabel('Team')
plt.ylabel('NRR')
for i, row in sorted_df.iterrows():
    plt.text(row['TEAM'], row['NRR'] + 0.01, round(row['NRR'], 2), ha='center')
plt.tight_layout()
plt.show()
```

```python
sorted_df = table.sort_values(by='PTS')
plt.figure(figsize=(12, 6))
plt.plot(sorted_df['TEAM'], sorted_df['PTS'], marker='o', linestyle='-', color='green')
plt.title('Points (PTS) of Teams - Season 25')
plt.xlabel('Team')
plt.ylabel('PTS')
for i, row in sorted_df.iterrows():
    plt.text(row['TEAM'], row['PTS'], round(row['PTS'], 2), ha='center')
plt.tight_layout()
plt.show()
```
