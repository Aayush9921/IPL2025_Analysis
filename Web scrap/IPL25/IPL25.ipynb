```python
import numpy as np
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
driver.get('https://www.iplt20.com/stats/2025')
time.sleep(5)
view=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[4]/div/section/div/div[3]/div[2]/div[2]/div/div[3]/div/div[1]/div/a')
driver.execute_script('arguments[0].click()',view)
time.sleep(5)
html=driver.page_source
driver.quit()

```

```python
df=pd.read_html(html)
len(df)
```

```python
stats=df[0]
stats.head()
```

```python
stats['Team'] = stats['Player'].apply(lambda x: x.split()[-1])
stats['Player'] = stats['Player'].apply(lambda x: ' '.join(x.split()[:-1]))

team_col = stats.pop('Team')
stats.insert(2, 'Team', team_col)

print(stats.head())
```

```python
stats.head()
```

```python
stats.to_csv('Statics.csv',index=False)
```

```python
import requests
r=requests.get('https://www.iplt20.com/auction')
s=BeautifulSoup(r.text,'lxml')
l=[]
f=s.find_all('span',class_='fr-fund')
for i in f:
  n=i.text
  l.append(n)
print(len(l))
```

```python
l=np.array(l).reshape(10,3)
print(l)
```

```python
team = ["CSK","DC","GT","KKR","LSG","MI","PBKS","RR","RCB","SRH"]
df=pd.DataFrame(
  np.hstack((np.array(team).reshape(10,1),l))
  ,columns=['Team','Fund_Remaining(Rs.)','overseas_plaayer','Total_player'])
```

```python
df.to_csv('Auction_overview.csv')
```

```python
driver=webdriver.Chrome()
driver.get('https://www.iplt20.com/auction')
time.sleep(3)
elem=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/section/div[2]/section/div[1]/ul/li[3]/a')
driver.execute_script('arguments[0].click()',elem)
html=driver.page_source
driver.quit()
```

```python
df=pd.read_html(html)
print(len(df))
```

```python
dfs=[df[1],df[2],df[3],df[4],df[5],df[6],df[7],df[8],df[9],df[10]]

teams=['CSK','DC','GT','KKR','LSG','MI','PBKS','RR','RCB','SRH']

for df,team in zip(dfs,teams):
  df['Team']=team

Sold_players=pd.concat(dfs,ignore_index=True)
Sold_players=Sold_players.drop(['Sr. No.','Unnamed: 5'],axis=1)
Sold_players=Sold_players[['Team','Player','Base Price','Winning Bid','Capped/Uncapped']]
Sold_players
```

```python
Sold_players.to_csv('Sold_players.csv')
```
